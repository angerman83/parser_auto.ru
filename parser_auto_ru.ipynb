```python
import pandas as pd
import requests
from datetime import datetime
```

```python
# функция получения данных от сервера
def fetch (url, params):
    headers = params['headers']
    body = params['body']
    method = params['method']
    if method == 'POST':
        return requests.post(url, headers=headers, data=body)
    if method == 'GET':
        return requests.get(url, headers=headers)
```

```python
# функция обработки полученных данных
def parse_item(car):
    c = {}
    c['availability'] = car['availability']
    c['owners_number'] = car['documents']['owners_number']
    try:
        c['pts_original'] = car['documents']['pts_original']
    except:
        pass
    try:
        c['custom_cleared'] = car['documents']['custom_cleared']
    except:
        pass
    c['documents_year'] = car['documents']['year']
    try:
        c['pts'] = car['documents']['pts']
    except:
        pass
    c['status'] = car['status']
    c['hidden'] = car['additional_info']['hidden']
    c['is_on_moderation'] = car['additional_info']['is_on_moderation']
    c['update_date'] = car['additional_info']['update_date']
    c['creation_date'] = car['additional_info']['creation_date']
    c['fresh_date'] = car['additional_info']['fresh_date']
    c['days_in_stock'] = car['additional_info']['days_in_stock']
    c['days_on_sale'] = car['additional_info']['days_on_sale']
    c['can_view'] = car['additional_info']['can_view']
    c['displayed_creation_date'] = car['additional_info']['displayed_creation_date']
    c['category'] = car['category']
    c['counters_chats_all'] = car['counters']['chats_all']
    c['counters_card_view_chat_conversion_all'] = car['counters']['card_view_chat_conversion_all']
    c['id'] = car['id']
    c['price_RUR'] = car['price_info']['RUR']
    c['price_EUR'] = car['price_info']['EUR']
    c['price_USD'] = car['price_info']['USD']
    c['price_with_nds'] = car['price_info']['with_nds']
    c['price_history'] = car['price_history']
    c['saleId'] = car['saleId']
    c['salon_is_oficial'] = car['salon']['is_oficial']
    c['section'] = car['section']
    c['seller_region_id'] = car['seller']['location']['region_info']['id']
    c['seller_region_name'] = car['seller']['location']['region_info']['name']
    c['seller_region_parent_ids'] = car['seller']['location']['region_info']['parent_ids']
    try:
        c['seller_name'] = car['seller']['name']
    except:
        pass
    c['seller_type'] = car['seller_type']
    c['condition'] = car['state']['condition']
    c['mileage'] = car['state']['mileage']
    pics = []
    for x in car['state']['image_urls']:
        pics.append(x['sizes']['1200x900'])
    c['images'] = pics
    c['sub_category'] = car['sub_category']
    try:
        c['axis'] = car['vehicle_info']['axis']
    except:
        pass
    try:
        c['brakes'] = car['vehicle_info']['brakes']
    except:
        pass
    try:
        c['loading'] = car['vehicle_info']['loading']
    except:
        pass
    c['mark_code'] = car['vehicle_info']['mark_info']['code']
    c['mark_name'] = car['vehicle_info']['mark_info']['name']
    try:
        c['mark_id'] = car['vehicle_info']['mark_info']['numeric_id']
    except:
        pass
    c['model_code'] = car['vehicle_info']['model_info']['code']
    c['model_name'] = car['vehicle_info']['model_info']['name']
    try:
        c['suspension'] = car['vehicle_info']['suspension']
    except:
        pass
    for x in car['vehicle_info']['equipment']:
        c['equipment_'+x] = car['vehicle_info']['equipment'][x]
    try:
        c['trailer_type'] = car['vehicle_info']['trailer_type']
    except:
        pass
    c['truck_category'] = car['vehicle_info']['truck_category']
    c['trucks_category'] = car['vehicle_info']['trucks_category']
    c['title'] = car['title']
    c['url'] = car['url']
    return c
```
